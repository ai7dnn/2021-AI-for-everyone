```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_sample_images

dataset = load_sample_images()   
img_rgb = dataset.images[1]
img_rgb.shape
```

```python
type(dataset.images)
```

```python
len(dataset.images)
```

```python
plt.figure(figsize=(8, 6))

plt.imshow(img_rgb)
plt.axis("off")
plt.title("RGB 이미지")
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(141)
plt.imshow(img_rgb[:, :, :])
plt.axis("off")
plt.title("RGB 이미지")

plt.subplot(142)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.Reds) # 마지막 첨자 = 0, 붉은색 강도
plt.axis("off")
plt.title("R 채널")

plt.subplot(143)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.Greens) # 마지막 첨자 = 1, 녹색 강도
plt.title("G 채널")

plt.subplot(144)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.Blues) # 마지막 첨자 = 2, 파란색 강도
plt.axis("off")
plt.title("B 채널")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
