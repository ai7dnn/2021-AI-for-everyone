```python
# 한글 폰트를 선명하도록 하는 파워명령어
%config InlineBackend.figure_format = 'retina'

import matplotlib as mpl
import matplotlib.pyplot as plt

# 한글 폰트 지원
plt.rc('font', family='NanumBarunGothic') 
# 음수 부호 지원
mpl.rcParams['axes.unicode_minus'] = False

import numpy as np
import matplotlib.pyplot as plt

plt.plot(range(-20, 80), np.random.randn(100).cumsum(), 'r--')
plt.title('한글 점검');
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(141)
# plt.imshow(img_rgb[50:200, 50:200, :])
plt.imshow(img_rgb[:, :, :])
plt.axis("off")
plt.title("RGB 이미지")

plt.subplot(142)
# plt.imshow(img_rgb[50:200, 50:200, 0], cmap=plt.cm.bone)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.Reds)
plt.axis("off")
plt.title("R 채널")

plt.subplot(143)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.Greens)
plt.title("G 채널")

plt.subplot(144)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.Blues)
# plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.bone)
plt.imshow(img_rgb[:, :, 2])
plt.axis("off")
plt.title("B 채널")

plt.show()
```
